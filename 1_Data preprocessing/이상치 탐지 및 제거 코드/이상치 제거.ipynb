```python
import pandas as pd

# 데이터 로드
DJ_all = pd.read_excel("수정_대전 물가 데이터.xlsx")

# '조사기준' 컬럼 제거 (필요하면 사용)
if '조사기준' in DJ_all.columns:
    DJ_all = DJ_all.drop(columns=['조사기준'])

# '월' 컬럼을 datetime 형식으로 변환
DJ_all['월'] = pd.to_datetime(DJ_all['월'], format='%Y-%m')

# 원본 데이터 복사 (이상치 제거 전 데이터 보존)
original_DJ_all = DJ_all.copy()

# 기간별 그룹화 설정
time_ranges = {
    "2014-2016": (2014, 2016),
    "2017-2019": (2017, 2019),
    "2020-2022": (2020, 2022),
    "2023-2024": (2023, 2024)
}

# 이상치 제거 함수 (IQR 방식)
def remove_outliers(group):
    # 수치형 데이터만 선택
    numeric_cols = group.select_dtypes(include=['number'])
    
    # IQR 계산
    Q1 = numeric_cols.quantile(0.25)
    Q3 = numeric_cols.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # 이상치 저장
    outlier_mask = (numeric_cols < lower_bound) | (numeric_cols > upper_bound)
    removed_outliers = numeric_cols[outlier_mask].stack().reset_index()
    removed_outliers.columns = ['index', '동', '제거된 값']

    # 이상치 처리
    cleaned_group = group.copy()
    for col in numeric_cols.columns:
        cleaned_group[col] = cleaned_group[col].where((cleaned_group[col] >= lower_bound[col]) & (cleaned_group[col] <= upper_bound[col]))

    return cleaned_group, removed_outliers

# 이상치 제거 적용
cleaned_dfs = []
removed_outliers_list = []

for label, (start_year, end_year) in time_ranges.items():
    # 해당 기간 데이터 필터링
    subset = DJ_all[(DJ_all['월'].dt.year >= start_year) & (DJ_all['월'].dt.year <= end_year)].copy()
    
    # 품목별, 지역별 이상치 제거
    cleaned_subset_list = []
    removed_outliers_list_per_period = []

    for _, group in subset.groupby('품목'):
        cleaned_group, removed_outliers = remove_outliers(group)
        cleaned_subset_list.append(cleaned_group)
        removed_outliers_list_per_period.append(removed_outliers)

    # 정제된 데이터와 제거된 이상치 데이터 합치기
    cleaned_dfs.append(pd.concat(cleaned_subset_list, ignore_index=True))
    removed_outliers_list.append(pd.concat(removed_outliers_list_per_period, ignore_index=True))

# 이상치 제거된 데이터 병합
cleaned_DJ_all = pd.concat(cleaned_dfs, ignore_index=True)
removed_outliers_df = pd.concat(removed_outliers_list, ignore_index=True)

# 이상치 제거 개수 집계
removed_outliers_count = removed_outliers_df.groupby('동')['제거된 값'].count().reset_index()
removed_outliers_count.columns = ['동', '제거된 개수']

# 결과 저장
cleaned_DJ_all.to_excel("대전_물가_IQR_이상치제거.xlsx", index=False)
removed_outliers_df.to_excel("이상치_제거_목록.xlsx", index=False)
removed_outliers_count.to_excel("이상치_제거_개수.xlsx", index=False)

# 결과 출력
print("✅ 이상치 제거된 데이터 저장 완료: '대전_물가_IQR_이상치제거.xlsx'")
print("📄 이상치 제거 개수 저장 완료: '이상치_제거_개수.xlsx'")
print("🗂 이상치 제거 목록 저장 완료: '이상치_제거_목록.xlsx'")
print("\n🔍 [이상치 제거 개수 상위 10개 동]")
print(removed_outliers_count.sort_values(by='제거된 개수', ascending=False).head(10))

print("\n🔍 [이상치 제거 목록 상위 10개]")
print(removed_outliers_df.head(10))
```

```python
import pandas as pd

# 데이터 로드
DJ_all = pd.read_excel("수정_대전 물가 데이터.xlsx")

# '조사기준' 컬럼 제거 (필요하면 사용)
if '조사기준' in DJ_all.columns:
    DJ_all = DJ_all.drop(columns=['조사기준'])

# '월' 컬럼을 datetime 형식으로 변환
DJ_all['월'] = pd.to_datetime(DJ_all['월'], format='%Y-%m')

# 원본 데이터 복사 (이상치 제거 전 데이터 보존)
original_DJ_all = DJ_all.copy()

# 기간별 그룹화 설정
time_ranges = {
    "2014-2016": (2014, 2016),
    "2017-2019": (2017, 2019),
    "2020-2022": (2020, 2022),
    "2023-2024": (2023, 2024)
}

# 이상치 제거 함수 (IQR 방식)
def remove_outliers(group):
    # 수치형 데이터만 선택
    numeric_cols = group.select_dtypes(include=['number'])
    
    # IQR 계산
    Q1 = numeric_cols.quantile(0.25)
    Q3 = numeric_cols.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # 이상치 저장
    outlier_mask = (numeric_cols < lower_bound) | (numeric_cols > upper_bound)
    removed_outliers = numeric_cols[outlier_mask].stack().reset_index()
    removed_outliers.columns = ['index', '동', '제거된 값']

    # 🔹 '품목'과 '월' 정보를 안전하게 추가
    removed_outliers = removed_outliers.merge(group[['월', '품목']], left_on='index', right_index=True, how='left')

    # 이상치 처리
    cleaned_group = group.copy()
    for col in numeric_cols.columns:
        cleaned_group[col] = cleaned_group[col].where((cleaned_group[col] >= lower_bound[col]) & (cleaned_group[col] <= upper_bound[col]))

    return cleaned_group, removed_outliers

# 이상치 제거 적용
cleaned_dfs = []
removed_outliers_list = []

for label, (start_year, end_year) in time_ranges.items():
    # 해당 기간 데이터 필터링
    subset = DJ_all[(DJ_all['월'].dt.year >= start_year) & (DJ_all['월'].dt.year <= end_year)].copy()
    
    # 품목별, 지역별 이상치 제거
    cleaned_subset_list = []
    removed_outliers_list_per_period = []

    for _, group in subset.groupby('품목'):
        cleaned_group, removed_outliers = remove_outliers(group)
        cleaned_subset_list.append(cleaned_group)
        removed_outliers_list_per_period.append(removed_outliers)

    # 정제된 데이터와 제거된 이상치 데이터 합치기
    cleaned_dfs.append(pd.concat(cleaned_subset_list, ignore_index=True))
    removed_outliers_list.append(pd.concat(removed_outliers_list_per_period, ignore_index=True))

# 이상치 제거된 데이터 병합
cleaned_DJ_all = pd.concat(cleaned_dfs, ignore_index=True)
removed_outliers_df = pd.concat(removed_outliers_list, ignore_index=True)

# ✅ 동별 & 품목별 제거된 이상치 개수 집계
removed_outliers_summary = removed_outliers_df.groupby(['동', '품목'])['제거된 값'].count().reset_index()
removed_outliers_summary.columns = ['동', '품목', '제거된 개수']

# ✅ 이상치 제거 개수 집계 (동별)
removed_outliers_count = removed_outliers_df.groupby('동')['제거된 값'].count().reset_index()
removed_outliers_count.columns = ['동', '제거된 개수']

# ✅ 이상치 제거된 상세 목록 저장
removed_outliers_detail = removed_outliers_df[['월', '동', '품목', '제거된 값']]

# 결과 저장
cleaned_DJ_all.to_excel("대전_물가_IQR_이상치제거.xlsx", index=False)
removed_outliers_df.to_excel("이상치_제거_목록.xlsx", index=False)
removed_outliers_count.to_excel("이상치_제거_개수.xlsx", index=False)
removed_outliers_summary.to_excel("이상치_제거_요약.xlsx", index=False)
removed_outliers_detail.to_excel("이상치_제거_상세목록.xlsx", index=False)

# 결과 출력
print("✅ 이상치 제거된 데이터 저장 완료: '대전_물가_IQR_이상치제거.xlsx'")
print("📄 이상치 제거 개수 저장 완료: '이상치_제거_개수.xlsx'")
print("📊 동별 & 품목별 이상치 개수 저장 완료: '이상치_제거_요약.xlsx'")
print("🗂 이상치 제거 목록 저장 완료: '이상치_제거_목록.xlsx'")
print("📅 이상치 제거된 월 포함한 목록 저장 완료: '이상치_제거_상세목록.xlsx'")

print("\n🔍 [동별 & 품목별 이상치 제거 개수 상위 10개]")
print(removed_outliers_summary.sort_values(by='제거된 개수', ascending=False).head(10))

print("\n📅 [이상치 제거된 월 정보 상위 10개]")
print(removed_outliers_detail.head(10))
```
