```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import AutoDateLocator, DateFormatter
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# 날짜 데이터 변환 및 정렬
df['시간'] = pd.to_datetime(df['시간'])
df = df.sort_values(by=['품목', '동', '시간'])

# '김' 품목만 필터링
df_egg = df[df['품목'] == '김']

# 이전 3개월 평균 가격 계산 (인덱스 정리)
df_egg['이전_3개월_평균'] = df_egg.groupby(['동'])['가격'].transform(lambda x: x.rolling(window=3).mean())

# 결측치 제거 (초반 3개월 데이터는 평균값이 없으므로)
df_egg = df_egg.dropna()

# 계절성 지수(Seasonality Index) 추가
monthly_avg = df_egg.groupby(df_egg['시간'].dt.month)['가격'].mean()
df_egg['계절성_지수'] = df_egg['시간'].dt.month.map(monthly_avg)

# 결과 저장용 딕셔너리
results = {}

# 동 별로 모델 생성 및 예측
dong_list = df_egg['동'].unique()
for dong in dong_list:
    df_dong = df_egg[df_egg['동'] == dong]

    # 시간순으로 데이터 분할 (70% 학습, 30% 검증)
    split_index = int(len(df_dong) * 0.7)
    train_data = df_dong.iloc[:split_index].copy()
    test_data = df_dong.iloc[split_index:].copy()

    X_train, y_train = train_data[['이전_3개월_평균']], train_data['가격']

    # 선형 회귀 모델 학습
    model = LinearRegression()
    model.fit(X_train, y_train)

    # 재귀적 예측을 수행하기 위한 리스트
    predicted_prices = []
    past_prices = list(train_data['가격'].iloc[-3:])  # 학습 데이터의 마지막 3개 가격 사용

    for i in range(len(test_data)):
        # 최근 3개월 평균 계산
        if len(past_prices) >= 3:
            moving_avg = np.mean(past_prices[-3:])
        else:
            moving_avg = np.mean(past_prices)  # 초기 데이터 부족 시

        # 새로운 데이터를 사용해 예측
        pred_price = model.predict(np.array([[moving_avg]]))[0]

        # 예측값 저장 및 리스트 업데이트
        predicted_prices.append(pred_price)
        past_prices.append(pred_price)  # 새로운 예측값을 추가하여 이동 평균 갱신

    # 예측값을 pandas Series로 변환하여 인덱스 맞추기
    y_pred_series = pd.Series(predicted_prices, index=test_data.index)

    # 성능 평가
    mae = mean_absolute_error(test_data['가격'], y_pred_series)
    rmse = np.sqrt(mean_squared_error(test_data['가격'], y_pred_series))

    # 선형 회귀 모델의 계수 및 절편 출력
    coefficients = model.coef_
    intercept = model.intercept_

    # p-value 계산을 위해 statsmodels 사용
    X_train_sm = sm.add_constant(X_train)  # 상수항 추가
    ols_model = sm.OLS(y_train, X_train_sm).fit()
    p_values = ols_model.pvalues

    results[dong] = {
        'y_test': test_data['가격'],
        'y_pred': y_pred_series,
        'mae': mae,
        'rmse': rmse,
        'coefficients': coefficients,
        'intercept': intercept,
        'p_values': p_values
    }

    # 성능 출력
    print(f"Performance for 김 in {dong}:")
    print(f"  MAE: {mae:.2f}")
    print(f"  RMSE: {rmse:.2f}")
    print(f"  Coefficients: {coefficients}")
    print(f"  Intercept: {intercept}")
    print(f"  P-values: {p_values}")
    print("----------------------------------")

    # 시각화 (기존 데이터 + 예측값 포함 - 검증 세트)
    plt.figure(figsize=(10, 5))
    plt.plot(test_data['시간'], test_data['가격'].values, label='Actual Prices', marker='o', linestyle='solid', color='blue')
    plt.plot(test_data['시간'], y_pred_series.values, label='Predicted Prices', linestyle='dashed', marker='x', color='orange')

    ax = plt.gca()
    ax.xaxis.set_major_locator(AutoDateLocator())
    ax.xaxis.set_major_formatter(DateFormatter('%Y-%m'))
    plt.xticks(rotation=45)

    plt.title(f'Price Prediction for 김 in {dong} (Validation Set)')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.grid()
    plt.show()

    # 전체 데이터셋에 대한 예측 시각화
    plt.figure(figsize=(10, 5))
    plt.plot(df_dong['시간'], df_dong['가격'], label='Actual Prices', marker='o', linestyle='solid', color='blue')
    plt.plot(test_data['시간'], y_pred_series.values, label='Predicted Prices', linestyle='dashed', marker='x', color='orange')

    ax = plt.gca()
    ax.xaxis.set_major_locator(AutoDateLocator())
    ax.xaxis.set_major_formatter(DateFormatter('%Y-%m'))
    plt.xticks(rotation=45)

    plt.title(f'Price Prediction for 달갈 in {dong} (Full Dataset)')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.grid()
    plt.show()

# 결과 출력
print("Price prediction model for 김 has been trained, validated, and visualized correctly.")

```
