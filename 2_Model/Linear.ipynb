```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import AutoDateLocator, DateFormatter
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# ÎÇ†Ïßú Îç∞Ïù¥ÌÑ∞ Î≥ÄÌôò Î∞è Ï†ïÎ†¨
df['ÏãúÍ∞Ñ'] = pd.to_datetime(df['ÏãúÍ∞Ñ'])
df = df.sort_values(by=['ÌíàÎ™©', 'Îèô', 'ÏãúÍ∞Ñ'])

# 'ÍπÄ' ÌíàÎ™©Îßå ÌïÑÌÑ∞ÎßÅ
df_egg = df[df['ÌíàÎ™©'] == 'ÍπÄ']

# Ïù¥Ï†Ñ 3Í∞úÏõî ÌèâÍ∑† Í∞ÄÍ≤© Í≥ÑÏÇ∞ (Ïù∏Îç±Ïä§ Ï†ïÎ¶¨)
df_egg['Ïù¥Ï†Ñ_3Í∞úÏõî_ÌèâÍ∑†'] = df_egg.groupby(['Îèô'])['Í∞ÄÍ≤©'].transform(lambda x: x.rolling(window=3).mean())

# Í≤∞Ï∏°Ïπò Ï†úÍ±∞ (Ï¥àÎ∞ò 3Í∞úÏõî Îç∞Ïù¥ÌÑ∞Îäî ÌèâÍ∑†Í∞íÏù¥ ÏóÜÏúºÎØÄÎ°ú)
df_egg = df_egg.dropna()

# Í≥ÑÏ†àÏÑ± ÏßÄÏàò(Seasonality Index) Ï∂îÍ∞Ä
monthly_avg = df_egg.groupby(df_egg['ÏãúÍ∞Ñ'].dt.month)['Í∞ÄÍ≤©'].mean()
df_egg['Í≥ÑÏ†àÏÑ±_ÏßÄÏàò'] = df_egg['ÏãúÍ∞Ñ'].dt.month.map(monthly_avg)

# Í≤∞Í≥º Ï†ÄÏû•Ïö© ÎîïÏÖîÎÑàÎ¶¨
results = {}

# Îèô Î≥ÑÎ°ú Î™®Îç∏ ÏÉùÏÑ± Î∞è ÏòàÏ∏°
dong_list = df_egg['Îèô'].unique()
for dong in dong_list:
    df_dong = df_egg[df_egg['Îèô'] == dong]

    # ÏãúÍ∞ÑÏàúÏúºÎ°ú Îç∞Ïù¥ÌÑ∞ Î∂ÑÌï† (70% ÌïôÏäµ, 30% Í≤ÄÏ¶ù)
    split_index = int(len(df_dong) * 0.7)
    train_data = df_dong.iloc[:split_index].copy()
    test_data = df_dong.iloc[split_index:].copy()

    X_train, y_train = train_data[['Ïù¥Ï†Ñ_3Í∞úÏõî_ÌèâÍ∑†']], train_data['Í∞ÄÍ≤©']

    # ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏ ÌïôÏäµ
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Ïû¨Í∑ÄÏ†Å ÏòàÏ∏°ÏùÑ ÏàòÌñâÌïòÍ∏∞ ÏúÑÌïú Î¶¨Ïä§Ìä∏
    predicted_prices = []
    past_prices = list(train_data['Í∞ÄÍ≤©'].iloc[-3:])  # ÌïôÏäµ Îç∞Ïù¥ÌÑ∞Ïùò ÎßàÏßÄÎßâ 3Í∞ú Í∞ÄÍ≤© ÏÇ¨Ïö©

    for i in range(len(test_data)):
        # ÏµúÍ∑º 3Í∞úÏõî ÌèâÍ∑† Í≥ÑÏÇ∞
        if len(past_prices) >= 3:
            moving_avg = np.mean(past_prices[-3:])
        else:
            moving_avg = np.mean(past_prices)  # Ï¥àÍ∏∞ Îç∞Ïù¥ÌÑ∞ Î∂ÄÏ°± Ïãú

        # ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞Î•º ÏÇ¨Ïö©Ìï¥ ÏòàÏ∏°
        pred_price = model.predict(np.array([[moving_avg]]))[0]

        # ÏòàÏ∏°Í∞í Ï†ÄÏû• Î∞è Î¶¨Ïä§Ìä∏ ÏóÖÎç∞Ïù¥Ìä∏
        predicted_prices.append(pred_price)
        past_prices.append(pred_price)  # ÏÉàÎ°úÏö¥ ÏòàÏ∏°Í∞íÏùÑ Ï∂îÍ∞ÄÌïòÏó¨ Ïù¥Îèô ÌèâÍ∑† Í∞±Ïã†

    # ÏòàÏ∏°Í∞íÏùÑ pandas SeriesÎ°ú Î≥ÄÌôòÌïòÏó¨ Ïù∏Îç±Ïä§ ÎßûÏ∂îÍ∏∞
    y_pred_series = pd.Series(predicted_prices, index=test_data.index)

    # ÏÑ±Îä• ÌèâÍ∞Ä
    mae = mean_absolute_error(test_data['Í∞ÄÍ≤©'], y_pred_series)
    rmse = np.sqrt(mean_squared_error(test_data['Í∞ÄÍ≤©'], y_pred_series))

    # ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏Ïùò Í≥ÑÏàò Î∞è Ï†àÌé∏ Ï∂úÎ†•
    coefficients = model.coef_
    intercept = model.intercept_

    # p-value Í≥ÑÏÇ∞ÏùÑ ÏúÑÌï¥ statsmodels ÏÇ¨Ïö©
    X_train_sm = sm.add_constant(X_train)  # ÏÉÅÏàòÌï≠ Ï∂îÍ∞Ä
    ols_model = sm.OLS(y_train, X_train_sm).fit()
    p_values = ols_model.pvalues

    results[dong] = {
        'y_test': test_data['Í∞ÄÍ≤©'],
        'y_pred': y_pred_series,
        'mae': mae,
        'rmse': rmse,
        'coefficients': coefficients,
        'intercept': intercept,
        'p_values': p_values
    }

    # ÏÑ±Îä• Ï∂úÎ†•
    print(f"Performance for ÍπÄ in {dong}:")
    print(f"  MAE: {mae:.2f}")
    print(f"  RMSE: {rmse:.2f}")
    print(f"  Coefficients: {coefficients}")
    print(f"  Intercept: {intercept}")
    print(f"  P-values: {p_values}")
    print("----------------------------------")

    # ÏãúÍ∞ÅÌôî (Í∏∞Ï°¥ Îç∞Ïù¥ÌÑ∞ + ÏòàÏ∏°Í∞í Ìè¨Ìï® - Í≤ÄÏ¶ù ÏÑ∏Ìä∏)
    plt.figure(figsize=(10, 5))
    plt.plot(test_data['ÏãúÍ∞Ñ'], test_data['Í∞ÄÍ≤©'].values, label='Actual Prices', marker='o', linestyle='solid', color='blue')
    plt.plot(test_data['ÏãúÍ∞Ñ'], y_pred_series.values, label='Predicted Prices', linestyle='dashed', marker='x', color='orange')

    ax = plt.gca()
    ax.xaxis.set_major_locator(AutoDateLocator())
    ax.xaxis.set_major_formatter(DateFormatter('%Y-%m'))
    plt.xticks(rotation=45)

    plt.title(f'Price Prediction for ÍπÄ in {dong} (Validation Set)')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.grid()
    plt.show()

    # Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ÏÖãÏóê ÎåÄÌïú ÏòàÏ∏° ÏãúÍ∞ÅÌôî
    plt.figure(figsize=(10, 5))
    plt.plot(df_dong['ÏãúÍ∞Ñ'], df_dong['Í∞ÄÍ≤©'], label='Actual Prices', marker='o', linestyle='solid', color='blue')
    plt.plot(test_data['ÏãúÍ∞Ñ'], y_pred_series.values, label='Predicted Prices', linestyle='dashed', marker='x', color='orange')

    ax = plt.gca()
    ax.xaxis.set_major_locator(AutoDateLocator())
    ax.xaxis.set_major_formatter(DateFormatter('%Y-%m'))
    plt.xticks(rotation=45)

    plt.title(f'Price Prediction for Îã¨Í∞à in {dong} (Full Dataset)')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.grid()
    plt.show()

# Í≤∞Í≥º Ï∂úÎ†•
print("Price prediction model for ÍπÄ has been trained, validated, and visualized correctly.")

```
